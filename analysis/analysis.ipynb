```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/trackFeatures.csv")
df.insert(loc=0, column='playlist name', value=df['playlistName'])
df.insert(loc=1, column='name', value=df['trackName'])
df.insert(loc=2, column='artist', value=df['trackArtist'])
df = df.drop(columns=['Unnamed: 0','trackName', 'trackArtist', 'playlistName'])
df
```

```python
fig = plt.figure(figsize = (20, 10))
playlists = df['playlist name'].unique() 
means = df.groupby('playlist name').mean()

ind = np.arange(len(playlists)) 
width = 0.25
plt.bar(ind - width, means['danceability'], width, label = 'Mean Danceability')
plt.bar(ind, means['energy'], width, label = 'Mean Energy')
plt.bar(ind + width, means['valence'], width, label = 'Mean Valence')
  
plt.xticks(ind, playlists)
plt.xlabel("Playlists")
plt.xticks(rotation = 90)
plt.title("Feature Analysis of the playlists")
plt.legend()
plt.show()
```
